```python
# Import Library
import pandas as pd
import numpy as np
```

```python
# Read/Import Data
data = pd.read_csv('https://raw.githubusercontent.com/densaiko/data_science_learning/main/dataset/insurance.csv')
data
```

# 1. Pengelompokkan BMI Category menggunakan Lambda Function

```python
data["bmi_category_lambda"] = data['bmi'].apply(
  lambda x: "High BMI" 
  if x>=50 
  else "Medium BMI"
  if x>=30 and x<50
  else "Low BMI")

data.sort_values(by=['bmi_category_lambda'])
```

# 2. Pivot Table Min, Max, dan Mean dari kolom "bmi_category_lambda" 

```python
table = pd.pivot_table(data, values=["bmi"], index=["bmi_category_lambda"], 
                       aggfunc={"bmi": [min, max, np.mean]})
table
```

# 3. Uppercase data pada kolom "sex", "region", dan "bmi_category_lambda"

```python
# Applying upper() method on '' column
data['sex'] = data['sex'].apply(lambda x: str.upper(x))
data['region'] = data['region'].apply(lambda x: str.upper(x))
data['bmi_category_lambda'] = data['bmi_category_lambda'].apply(lambda x: str.upper(x))
data.sort_values(by=['bmi_category_lambda'])
```
